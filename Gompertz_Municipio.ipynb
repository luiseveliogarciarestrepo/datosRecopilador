```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 500)
```

```python
gompertz = pd.read_csv('CovidColombia.csv', low_memory=False, usecols= ['ID de caso', 'Fecha de notificación', 
                                                                        'Ciudad de ubicación', 
                                                                        'Departamento o Distrito '], 
                                                                        parse_dates=['Fecha de notificación'] )
```

```python
gompertz.sort_values(by=['Fecha de notificación'],inplace=True)
```

```python
gompertz['Fecha de notificación']= gompertz['Fecha de notificación'].dt.floor('d')
```

```python
df = pd.DataFrame(gompertz['Fecha de notificación']).loc[0]
```

```python
a = gompertz.loc[0,'Fecha de notificación']
```

```python
longitud = (b-a)/ np.timedelta64(1, 'D')
longitud=int(longitud)-1
longitud
```

```python
municipio = 'Manizales'
```

```python
gompertz.columns
```

```python
filtro = gompertz['Ciudad de ubicación'] == municipio
gompertz_mun = gompertz.loc[filtro,:]
```

```python
gompertz_mun=gompertz_mun.append(df_fechas).sort_values(by=['Fecha de notificación'], ignore_index=True)
```

```python
gompertz_mun
```

```python
longitud = gompertz_mun['ID de caso'].size
longitud
```

```python
gompertz_mun.columns
```

```python
for i in range(0, longitud):
    if gompertz_mun.loc[i, 'ID de caso'] == 0:
        pass
    else:
        gompertz_mun.loc[i, 'ID de caso'] = 1
```

```python
gompertz_mun
```

```python
gompertz_mun['ID de caso'].sum()
```

```python
gompertz_mun['Acumulados']= 0
gompertz_mun['Día número']=0
gompertz_mun['Penetración real']=0
gompertz_mun
```

```python
gompertz_mun.loc[0, 'Acumulados']= gompertz_mun.loc[0, 'ID de caso']
for i in range(1, longitud):
    gompertz_mun.loc[i,'Acumulados'] = gompertz_mun.loc[i, 'ID de caso'] + gompertz_mun.loc[i-1, 'Acumulados']
```

```python
gompertz_mun
```

```python
gompertz_mun.columns
```

```python
gompertz_mun = gompertz_mun.groupby('Fecha de notificación')[['ID de caso', 'Ciudad de ubicación',
       'Departamento o Distrito ', 'Acumulados', 'Día número',
       'Penetración real']].max()
```

```python
gompertz_mun.reset_index(inplace=True)
```

```python
longitud = gompertz_mun['ID de caso'].size
longitud
```

```python
gompertz_mun.loc[0, 'Día número'] = 1
for i in range(1, longitud):
    gompertz_mun.loc[i,'Día número'] = i+1
    gompertz_mun.loc[i,'Penetración real'] = gompertz_mun.loc[i, 'Acumulados']/ gompertz_mun.loc[longitud-1, 'Acumulados']
    
```
